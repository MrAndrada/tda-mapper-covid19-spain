```python
def convertir_fecha_a_numero(fecha):
    fecha_base = datetime(2020, 1, 1)  # Fecha de referencia: 1 de enero de 2020
    fecha_objeto = datetime.strptime(fecha, '%Y-%m-%d')  # Convierte la fecha de cadena a objeto datetime
    delta = fecha_objeto - fecha_base
    return delta.days
```

```python
df_spain_ministerio_sanidad['Fecha'] = df_spain_ministerio_sanidad['Fecha'].apply(convertir_fecha_a_numero)
df_spain_ministerio_sanidad
```

```python
# Eliminar instancias donde la CCAA no consta
df_spain_ministerio_sanidad = df_spain_ministerio_sanidad[df_spain_ministerio_sanidad['CCAA'] != 'No consta']
df_spain_ministerio_sanidad
```

```python
# Extraer nombre de Comunidades Autónomas
comunidades_autonomas = df_spain_ministerio_sanidad['CCAA'].unique().tolist()
comunidades_autonomas
```

```python
def add_coordinates_to_regions(ccaa):

    # Crear un geocoder
    geolocator = Nominatim(user_agent="my_geocoder")

    # Lista para almacenar las coordenadas
    coordinates = []

    # Obtener el nombre de la Comunidad Autónoma
    comunidad = ccaa

    # Geolocalizar la Comunidad Autónoma
    location = geolocator.geocode(comunidad + ", Spain")

    # Extraer las coordenadas
    latitude = location.latitude
    longitude = location.longitude

    # Agregar las coordenadas a la lista
    coordinates.append((comunidad, latitude, longitude))

    # Convertir la lista de coordenadas en un DataFrame
    coordinates_df = pd.DataFrame(coordinates, columns=['CCAA', 'Latitude', 'Longitude'])

    return coordinates_df

df_coordenadas_ccaa = pd.DataFrame()

for item in range(len(comunidades_autonomas)):
  df_coordenadas_ccaa_temp = add_coordinates_to_regions(comunidades_autonomas[item])
  df_coordenadas_ccaa = df_coordenadas_ccaa.append(df_coordenadas_ccaa_temp, ignore_index=True)
```

```python
df_merged_spain = pd.merge(df_spain_ministerio_sanidad, df_coordenadas_ccaa, on='CCAA', how='left')
df_merged_spain
```

```python
from google.colab import files

# Eliminar columnas innecesarias
df_merged_spain = df_merged_spain.drop(['cod_ine', 'Hospitalizados', 'UCI'], axis=1)
df_merged_spain

# Convertir a fichero Excel
df_merged_spain.to_excel('covid19_data_spain_ministerio_sanidad.xlsx', index=False)

# Descargar fichero generado
files.download('covid19_data_spain_ministerio_sanidad.xlsx')
```
